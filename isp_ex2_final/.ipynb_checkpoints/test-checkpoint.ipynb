rice_encode: This function takes a single audio sample and encodes it using Rice coding. In your case:
Sample Rate: 44100
Number of audio samples: 501022
For each of the 501022 audio samples, rice_encode is used to compress it using a specific k_value.
encode_audio: This function is responsible for encoding the entire audio signal by splitting it into smaller blocks and applying rice_encode to each block. In your case:
Number of blocks: 489
Here's how it works:
The audio signal (501022 audio samples) is divided into 489 blocks. Each block contains multiple audio samples.
For each block, rice_encode is applied to each of its audio samples.
The result is a list of encoded blocks, where each block contains encoded versions of the audio samples it originally held.
The goal is to compress the entire audio signal into smaller blocks, making it more efficient to store and transmit while preserving the audio quality as defined by the k_value. After encoding, you have an encoded version of the audio.

```python
import os
from scipy.io import wavfile
import numpy as np
import bitarray
import matplotlib.pyplot as plt

def load_audio_file(file_path):
    try:
        sample_rate, audio_data = wavfile.read(file_path)
        # print("\nOriginal audio Information:")
        # print(f"Audio file '{file_path}' loaded successfully.")
        # print(f"Sample Rate: {sample_rate}")
        # print(f"Audio Data Shape: {audio_data.shape}")
        # num_samples = len(audio_data)  # Get the number of samples
        # print(f"Number of audio samples: {num_samples}")
        # if audio_data.shape[0] == 2:
        #     print("Audio is stereo.")
        # else:
        #     print("Audio is mono.")

        return sample_rate, audio_data
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' does not exist.")
        return None, None
    except Exception as e:
        print(f"An error occurred while loading the audio file: {str(e)}")
        return None, None

def split_audio_into_blocks(audio_data, block_size):
    num_blocks = len(audio_data) // block_size
    blocks = np.array_split(audio_data, num_blocks)
    return blocks

def rice_encode(value, k_value):
    q, r = divmod(value, 2**k_value)
    unary_code = '1' * q + '0'
    binary_code = format(r, '0' + str(k_value) + 'b')
    encoded_value = unary_code + binary_code
    return encoded_value

def rice_decode(encoded_value, k_value):
    idx = encoded_value.index('0')
    q = idx
    binary_code = encoded_value[idx + 1 : idx + 1 + k_value]
    r = int(binary_code, 2)
    decoded_value = q * 2**k_value + r
    return decoded_value

def encode_audio(audio_data, k_value):
    encoded_blocks = []
    for block in audio_data:
        encoded_block = [rice_encode(value, k_value) for value in block]
        encoded_blocks.append(encoded_block)
    return encoded_blocks



def decode_audio(encoded_data, k_value):
    decoded_audio = []
    for encoded_block in encoded_data:
        decoded_block = []
        for encoded_value in encoded_block:
            # Convert encoded_value to a binary string
            encoded_value_str = str(encoded_value)
            
            # Extract the unary part (variable length) until '0' is encountered
            unary_length = 0
            while unary_length < len(encoded_value_str) and encoded_value_str[unary_length] == '1':
                unary_length += 1
            
            # Extract the binary code if available
            binary_code = encoded_value_str[unary_length + 1 : unary_length + 1 + k_value] if unary_length + 1 + k_value <= len(encoded_value_str) else ''
            
            q = len(encoded_value_str[:unary_length])  # Calculate 'q' from the unary code
            r = int(binary_code, 2) if binary_code else 0  # Handle empty binary code as zero
            decoded_value = q * 2**k_value + r
            decoded_block.append(decoded_value)
        
        # print(f"Decoded Block: {decoded_block}")  # Debugging line
        decoded_audio.append(decoded_block)
    
    return np.array(decoded_audio).flatten()



def check_encoded_data(encoded_file_path, k_value):
    try:
        with open(encoded_file_path, 'rb') as encoded_file:
            bit_stream = bitarray.bitarray()
            bit_stream.fromfile(encoded_file)
    except FileNotFoundError:
        print(f"Error: The encoded file '{encoded_file_path}' does not exist.")
        return None

    print(f"Bitstream Length: {len(bit_stream)}")

    # Decode audio data block by block
    decoded_audio_blocks = []
    bit_stream_index = 0

    # Decode audio data block by block using the decode_audio function
    decoded_audio_data = decode_audio(bit_stream, k_value)

    return decoded_audio_data

def visualize_audio_waveform(decoded_audio_data):
    if decoded_audio_data is not None:
        plt.figure(figsize=(10, 4))
        plt.plot(decoded_audio_data)
        plt.title("Decoded Audio Waveform")
        plt.xlabel("Sample")
        plt.ylabel("Amplitude")
        # plt.show()

def visualize_audio_blocks(audio_data, audio_blocks, block_size):
    plt.figure(figsize=(12, 6))
    plt.plot(audio_data, label='Original Audio')
    plt.title('Original Audio Data')
    plt.xlabel('Sample')
    plt.ylabel('Amplitude')

    # Plot vertical lines to represent block boundaries
    for i in range(1, len(audio_blocks)):
        plt.axvline(i * block_size, color='r', linestyle='--', linewidth=0.5)
        
    plt.legend()
    # plt.show()
    
    
def visualize_bitstream(bit_stream):
    plt.figure(figsize=(12, 6))
    plt.plot(bit_stream.tolist(), linestyle='-', marker='', markersize=2)
    plt.title('Encoded Bitstream')
    plt.xlabel('Bit Index')
    plt.ylabel('Bit Value')
    plt.show()

# def visualize_audio_difference(original_audio, decoded_audio_blocks, block_size):
#     plt.figure(figsize=(12, 6))
    
#     # Calculate the total number of samples
#     total_samples = len(original_audio)
    
#     # Calculate the x-axis values for plotting
#     x = np.arange(0, total_samples)
    
#     # Initialize an array to store the difference between original and decoded audio
#     difference = np.zeros(total_samples)
    
#     # Fill in the difference array using the decoded blocks
#     for i, block in enumerate(decoded_audio_blocks):
#         start_sample = i * block_size
#         end_sample = start_sample + len(block)
#         difference[start_sample:end_sample] = block - original_audio[start_sample:end_sample]
    
#     plt.plot(x, difference)
#     plt.title('Difference Between Original and Decoded Audio')
#     plt.xlabel('Sample')
#     plt.ylabel('Amplitude Difference')
#     plt.show()



    
def save_decoded_audio(decoded_audio_data, sample_rate, folder_name, file_name):
    if decoded_audio_data is not None:
        decoded_file_path = os.path.join(folder_name, file_name)
        try:
            wavfile.write(decoded_file_path, sample_rate, decoded_audio_data.astype(np.int16))
            print(f"Decoded audio data saved to '{decoded_file_path}'.")
        except Exception as e:
            print(f"An error occurred while saving the decoded audio data: {str(e)}")


def encode_and_save_audio(audio_blocks, k_value, folder_name, file_name):
    encoded_audio = encode_audio(audio_blocks, k_value)
    
    encoded_file_name = f"{os.path.splitext(file_name)[0]}_Enc.ex2"

    encoded_file_path = os.path.join(folder_name, encoded_file_name)

    save_encoded_data(encoded_audio, encoded_file_path)


def save_encoded_data(encoded_audio, encoded_file_path):
    with open(encoded_file_path, 'wb') as encoded_file:
        bit_stream = bitarray.bitarray()
        for encoded_block in encoded_audio:
            for encoded_value in encoded_block:
                bit_stream.extend(encoded_value)
        bit_stream.tofile(encoded_file)

    # print(f"Encoded data saved to '{encoded_file_path}'.")

def load_encoded_file(encoded_file_path):
    try:
        with open(encoded_file_path, 'rb') as encoded_file:
            bit_stream = bitarray.bitarray()
            bit_stream.fromfile(encoded_file)
        return bit_stream
    except FileNotFoundError:
        print(f"Error: The encoded file '{encoded_file_path}' does not exist.")
        return None
    except Exception as e:
        print(f"An error occurred while loading the encoded file: {str(e)}")
        return None


def decode_audio_blocks(bit_stream, audio_blocks, block_size, k_value):
    decoded_audio_blocks = []
    bit_stream_index = 0
    
    for _ in range(len(audio_blocks)):
        decoded_block = []
        
        for _ in range(block_size):
            encoded_value_str = bit_stream[bit_stream_index:bit_stream_index + (k_value + 1)].to01()
            encoded_value = int(encoded_value_str, 2)
            decoded_block.append(encoded_value)
            bit_stream_index += (k_value + 1)
        
        decoded_audio_blocks.append(decoded_block)
    
    return decoded_audio_blocks

def reconstruct_audio(decoded_audio_blocks, k_value):
    decoded_audio_data = decode_audio(decoded_audio_blocks, k_value)
    return decoded_audio_data

def save_decoded_audio(decoded_audio_data, sample_rate, folder_name, file_name):
    decoded_file_path = os.path.join(folder_name, file_name)
    
    try:
        wavfile.write(decoded_file_path, sample_rate, decoded_audio_data.astype(np.int16))
        print(f"Decoded audio data saved to '{decoded_file_path}'.")
    except Exception as e:
        print(f"An error occurred while saving the decoded audio data: {str(e)}")


def process_encoded_audio(bit_stream, audio_blocks, block_size, k_value, sample_rate, folder_name):
    print(f"Bitstream Length: {len(bit_stream)}")
    
    # Decode and reconstruct audio
    decoded_audio_blocks = decode_audio_blocks(bit_stream, audio_blocks, block_size, k_value)
    decoded_audio_data = reconstruct_audio(decoded_audio_blocks, k_value)
    
    # Save the decoded audio
    save_decoded_audio(decoded_audio_data, sample_rate, folder_name, 'Sound1_Enc_Dec.wav')

def check_sample_rate(bit_stream, audio_blocks, block_size, k_value):
    # Calculate the expected sample rate
    expected_sample_rate = len(bit_stream) / (len(audio_blocks) * block_size * (k_value + 1))
    
    return expected_sample_rate

def calculate_num_sample_rates(bit_stream, block_size, k_value):
    # Calculate the number of sample rates
    total_bits = len(bit_stream)
    bits_per_block = (k_value + 1) * block_size
    num_sample_rates = total_bits / bits_per_block
    return int(num_sample_rates)


# def main():
#     # Define file and folder names
#     folder_name = 'file'
#     file_name = 'Sound1.wav'
#     encoded_file_name = 'Sound1_Enc.ex2'

#     # Load audio file and get sample rate and audio data
#     sample_rate, audio_data = load_audio_file(os.path.join(folder_name, file_name))

#     if sample_rate is None or audio_data is None:
#         return  # Exit the function if loading fails

#     # Define encoding parameters
#     block_size = 1024  # You can adjust this based on your preferences
#     k_value = 4  # You can change this to 2 bits or any other value

#     # Split audio data into blocks
#     audio_blocks = split_audio_into_blocks(audio_data, block_size)

#     # Visualize the audio data split into blocks
#     visualize_audio_blocks(audio_data, audio_blocks, block_size)

#     # Encode and save audio data
#     encode_and_save_audio(audio_blocks, k_value, folder_name, file_name)

#     # Print the number of blocks
#     num_blocks = len(audio_blocks)
#     print(f"Number of blocks: {num_blocks}")

#     # Load the encoded file
#     encoded_file_path = os.path.join(folder_name, encoded_file_name)
#     bit_stream = load_encoded_file(encoded_file_path)


#     if bit_stream is not None:
#         # Check sample rate
#         if sample_rate != check_sample_rate(bit_stream, audio_blocks, block_size, k_value):
#             print("Sample rates do not match.")
#         else:
#             process_encoded_audio(bit_stream, audio_blocks, block_size, k_value, sample_rate, folder_name)
#     else:
#         print("Unable to load the encoded file.")
        

#     if bit_stream is not None:
#         decoded_audio_blocks = process_encoded_audio(bit_stream, audio_blocks, block_size, k_value, sample_rate, folder_name)

#         # Check for lossless encoding
#         if np.array_equal(audio_data, decoded_audio_blocks):
#             print("Encoding and decoding is lossless.")
#         else:
#             print("Encoding and decoding is not lossless. There may be artifacts.")


#         # Load and display information about the decoded audio file
#         decoded_file_name = 'Sound1_Enc_Dec.wav'
#         decoded_file_path = os.path.join(folder_name, decoded_file_name)
#         decoded_sample_rate, decoded_audio_data = load_audio_file(decoded_file_path)
        
#         if decoded_sample_rate is not None and decoded_audio_data is not None:
#             print(f"Decoded Audio file '{decoded_file_path}' loaded successfully.")
#             print(f"Sample Rate: {decoded_sample_rate}")
#             print(f"Decoded Audio Data Shape: {decoded_audio_data.shape}")
#             num_decoded_samples = len(decoded_audio_data)
#             print(f"Number of decoded audio samples: {num_decoded_samples}")
#         else:
#             print("Unable to load the decoded audio file.")

# if __name__ == "__main__":
#     main()






def main():
    # Define file and folder names
    folder_name = 'file'
    file_name = 'Sound1.wav'
    encoded_file_name = 'Sound1_Enc.ex2'

    # Load audio file and get sample rate and audio data
    sample_rate, audio_data = load_audio_file(os.path.join(folder_name, file_name))

    if sample_rate is None or audio_data is None:
        return  # Exit the function if loading fails

    # Define encoding parameters
    block_size = 1024  # You can adjust this based on your preferences
    k_value = 4  # You can change this to 2 bits or any other value

    # Split audio data into blocks
    audio_blocks = split_audio_into_blocks(audio_data, block_size)

    # Visualize the audio data split into blocks
    visualize_audio_blocks(audio_data, audio_blocks, block_size)

    # Encode and save audio data
    encode_and_save_audio(audio_blocks, k_value, folder_name, file_name)

    # Load the encoded file
    encoded_file_path = os.path.join(folder_name, encoded_file_name)
    bit_stream = load_encoded_file(encoded_file_path)

    # Print information about the original audio
    print("\nOriginal File Information:")
    print(f"Audio file '{file_name}' loaded successfully.")
    print(f"Sample Rate: {sample_rate}")
    print(f"Audio Data Shape: {audio_data.shape}")
    num_samples = len(audio_data)
    print(f"Number of audio samples: {num_samples}")
    if audio_data.shape[0] == 2:
        print("Audio is stereo.")
    else:
        print("Audio is mono.")

    # Print information about the encoded data
    print("\nEncoded File Information:")
    print(f"Encoded data saved to '{os.path.join(folder_name, encoded_file_name)}'.")
    print(f"Number of blocks: {len(audio_blocks)}")

    if bit_stream is not None:
       # Calculate the number of sample rates in the encoded data
        num_sample_rates = calculate_num_sample_rates(bit_stream, block_size, k_value)

        # Check sample rate
        if sample_rate != check_sample_rate(bit_stream, audio_blocks, block_size, k_value):
            print("Sample rates do not match.")
        else:
            process_encoded_audio(bit_stream, audio_blocks, block_size, k_value, sample_rate, folder_name)
    else:
        print("Unable to load the encoded file.")
        

    if bit_stream is not None:
        decoded_audio_blocks = process_encoded_audio(bit_stream, audio_blocks, block_size, k_value, sample_rate, folder_name)

        # Check for lossless encoding
        if np.array_equal(audio_data, decoded_audio_blocks):
            print("Encoding and decoding is lossless.")
        else:
            print("Encoding and decoding is not lossless. There may be artifacts.")


    # Load and display information about the decoded audio file
    decoded_file_name = 'Sound1_Enc_Dec.wav'
    decoded_file_path = os.path.join(folder_name, decoded_file_name)
    decoded_sample_rate, decoded_audio_data = load_audio_file(decoded_file_path)
    
    if decoded_sample_rate is not None and decoded_audio_data is not None:
        print("\nDecoded File Information:")
        print(f"Decoded Audio file '{decoded_file_name}' loaded successfully.")
        print(f"Sample Rate: {decoded_sample_rate}")
        print(f"Decoded Audio Data Shape: {decoded_audio_data.shape}")
        num_decoded_samples = len(decoded_audio_data)
        print(f"Number of decoded audio samples: {num_decoded_samples}")
    else:
        print("Unable to load the decoded audio file.")

if __name__ == "__main__":
    main()
```

```python
import os
from scipy.io import wavfile
import numpy as np
import bitarray
import matplotlib.pyplot as plt
from scipy.signal import resample 

def load_audio_file(file_path):
    try:
        sample_rate, audio_data = wavfile.read(file_path)
        return sample_rate, audio_data
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' does not exist.")
        return None, None
    except Exception as e:
        print(f"An error occurred while loading the audio file: {str(e)}")
        return None, None

def split_audio_into_blocks(audio_data, block_size):
    num_blocks = len(audio_data) // block_size
    blocks = np.array_split(audio_data, num_blocks)
    return blocks

def rice_encode(value, k_value):
    q, r = divmod(value, 2**k_value)
    unary_code = '1' * q + '0'
    binary_code = format(r, '0' + str(k_value) + 'b')
    encoded_value = unary_code + binary_code
    return encoded_value

def rice_decode(encoded_value, k_value):
    idx = encoded_value.index('0')
    q = idx
    binary_code = encoded_value[idx + 1 : idx + 1 + k_value]
    r = int(binary_code, 2)
    decoded_value = q * 2**k_value + r
    return decoded_value

# def encode_audio(audio_data, k_value):
#     encoded_blocks = []
#     for block in audio_data:
#         encoded_block = [rice_encode(value, k_value) for value in block]
#         encoded_blocks.append(encoded_block)
#     return encoded_blocks

def encode_audio(audio_data, k_value, block_size):
    encoded_blocks = []
    for block in audio_data:
        encoded_block = [rice_encode(value, k_value) for value in block]
        encoded_blocks.append(encoded_block)
    return encoded_blocks


def decode_audio(encoded_data, k_value):
    decoded_audio = []
    for encoded_block in encoded_data:
        decoded_block = []
        for encoded_value in encoded_block:
            # Convert encoded_value to a binary string
            encoded_value_str = str(encoded_value)
            
            # Extract the unary part (variable length) until '0' is encountered
            unary_length = 0
            while unary_length < len(encoded_value_str) and encoded_value_str[unary_length] == '1':
                unary_length += 1
            
            # Extract the binary code if available
            binary_code = encoded_value_str[unary_length + 1 : unary_length + 1 + k_value] if unary_length + 1 + k_value <= len(encoded_value_str) else ''
            
            q = len(encoded_value_str[:unary_length])  # Calculate 'q' from the unary code
            r = int(binary_code, 2) if binary_code else 0  # Handle empty binary code as zero
            decoded_value = q * 2**k_value + r
            decoded_block.append(decoded_value)
        
        # print(f"Decoded Block: {decoded_block}")  # Debugging line
        decoded_audio.append(decoded_block)
    
    return np.array(decoded_audio).flatten()



def check_encoded_data(encoded_file_path, k_value):
    try:
        with open(encoded_file_path, 'rb') as encoded_file:
            bit_stream = bitarray.bitarray()
            bit_stream.fromfile(encoded_file)
    except FileNotFoundError:
        print(f"Error: The encoded file '{encoded_file_path}' does not exist.")
        return None

    print(f"Bitstream Length: {len(bit_stream)}")

    # Decode audio data block by block
    decoded_audio_blocks = []
    bit_stream_index = 0

    # Decode audio data block by block using the decode_audio function
    decoded_audio_data = decode_audio(bit_stream, k_value)

    return decoded_audio_data

def visualize_audio_waveform(decoded_audio_data):
    if decoded_audio_data is not None:
        plt.figure(figsize=(10, 4))
        plt.plot(decoded_audio_data)
        plt.title("Decoded Audio Waveform")
        plt.xlabel("Sample")
        plt.ylabel("Amplitude")
        # plt.show()

def visualize_audio_blocks(audio_data, audio_blocks, block_size):
    plt.figure(figsize=(12, 6))
    plt.plot(audio_data, label='Original Audio')
    plt.title('Original Audio Data')
    plt.xlabel('Sample')
    plt.ylabel('Amplitude')

    # Plot vertical lines to represent block boundaries
    for i in range(1, len(audio_blocks)):
        plt.axvline(i * block_size, color='r', linestyle='--', linewidth=0.5)
        
    plt.legend()
    # plt.show()
    
    
def visualize_bitstream(bit_stream):
    plt.figure(figsize=(12, 6))
    plt.plot(bit_stream.tolist(), linestyle='-', marker='', markersize=2)
    plt.title('Encoded Bitstream')
    plt.xlabel('Bit Index')
    plt.ylabel('Bit Value')
    plt.show()
   
def save_decoded_audio(decoded_audio_data, sample_rate, folder_name, file_name):
    if decoded_audio_data is not None:
        decoded_file_path = os.path.join(folder_name, file_name)
        try:
            wavfile.write(decoded_file_path, sample_rate, decoded_audio_data.astype(np.int16))
            print(f"Decoded audio data saved to '{decoded_file_path}'.")
        except Exception as e:
            print(f"An error occurred while saving the decoded audio data: {str(e)}")


# def encode_and_save_audio(audio_blocks, k_value, folder_name, file_name):
#     encoded_audio = encode_audio(audio_blocks, k_value)
    
#     encoded_file_name = f"{os.path.splitext(file_name)[0]}_Enc.ex2"

#     encoded_file_path = os.path.join(folder_name, encoded_file_name)

#     save_encoded_data(encoded_audio, encoded_file_path)

def encode_and_save_audio(audio_blocks, k_value, block_size, folder_name, file_name):
    encoded_audio = encode_audio(audio_blocks, k_value, block_size)
    
    encoded_file_name = f"{os.path.splitext(file_name)[0]}_Enc.ex2"

    encoded_file_path = os.path.join(folder_name, encoded_file_name)

    save_encoded_data(encoded_audio, encoded_file_path)


    
def save_encoded_data(encoded_audio, encoded_file_path):
    with open(encoded_file_path, 'wb') as encoded_file:
        bit_stream = bitarray.bitarray()
        for encoded_block in encoded_audio:
            for encoded_value in encoded_block:
                bit_stream.extend(encoded_value)
        bit_stream.tofile(encoded_file)

    # print(f"Encoded data saved to '{encoded_file_path}'.")

def load_encoded_file(encoded_file_path):
    try:
        with open(encoded_file_path, 'rb') as encoded_file:
            bit_stream = bitarray.bitarray()
            bit_stream.fromfile(encoded_file)
        return bit_stream
    except FileNotFoundError:
        print(f"Error: The encoded file '{encoded_file_path}' does not exist.")
        return None
    except Exception as e:
        print(f"An error occurred while loading the encoded file: {str(e)}")
        return None


def decode_audio_blocks(bit_stream, audio_blocks, block_size, k_value):
    decoded_audio_blocks = []
    bit_stream_index = 0
    
    for _ in range(len(audio_blocks)):
        decoded_block = []
        
        for _ in range(block_size):
            encoded_value_str = bit_stream[bit_stream_index:bit_stream_index + (k_value + 1)].to01()
            encoded_value = int(encoded_value_str, 2)
            decoded_block.append(encoded_value)
            bit_stream_index += (k_value + 1)
        
        decoded_audio_blocks.append(decoded_block)
    
    return decoded_audio_blocks

def reconstruct_audio(decoded_audio_blocks, k_value):
    decoded_audio_data = decode_audio(decoded_audio_blocks, k_value)
    return decoded_audio_data

def save_decoded_audio(decoded_audio_data, sample_rate, folder_name, file_name):
    decoded_file_path = os.path.join(folder_name, file_name)
    
    try:
        wavfile.write(decoded_file_path, sample_rate, decoded_audio_data.astype(np.int16))
        print(f"Decoded audio data saved to '{decoded_file_path}'.")
    except Exception as e:
        print(f"An error occurred while saving the decoded audio data: {str(e)}")


# def process_encoded_audio(bit_stream, audio_blocks, block_size, k_value, sample_rate, folder_name, audio_data):
#     print(f"Encoded Bitstream Length: {len(bit_stream)}")
    
#     # Calculate the expected length of the decoded bitstream
#     expected_decoded_bitstream_length = len(audio_blocks) * block_size * (k_value + 1)
    
#     # Decode and reconstruct audio
#     decoded_audio_blocks = decode_audio_blocks(bit_stream, audio_blocks, block_size, k_value)
#     decoded_audio_data = reconstruct_audio(decoded_audio_blocks, k_value)
    
#     # Resample the decoded audio to match the original sample rate
#     decoded_audio_data_resampled = resample(decoded_audio_data, len(audio_data))
    
#     # Convert the data type to np.int16
#     decoded_audio_data_resampled = decoded_audio_data_resampled.astype(np.int16)
    
#     # Visualize the decoded audio
#     visualize_audio_waveform(decoded_audio_data_resampled)
    
#     # Save the resampled decoded audio
#     save_decoded_audio(decoded_audio_data_resampled, sample_rate, folder_name, 'Sound1_Enc_Dec.wav')

#     # Return the decoded bitstream
#     decoded_bit_stream = encode_audio(decoded_audio_blocks, k_value)

#     print(f"Expected Decoded Bitstream Length: {expected_decoded_bitstream_length}")
#     print(f"Actual Decoded Bitstream Length: {len(decoded_bit_stream)}")

#     return decoded_audio_data_resampled, decoded_bit_stream

def process_encoded_audio(bit_stream, audio_blocks, block_size, k_value, sample_rate, folder_name, audio_data):
    print(f"Encoded Bitstream Length: {len(bit_stream)}")
    
    # Calculate the expected length of the decoded bitstream
    expected_decoded_bitstream_length = len(audio_blocks) * block_size * (k_value + 1)
    
    # Decode and reconstruct audio using the provided block_size
    decoded_audio_blocks = decode_audio_blocks(bit_stream, audio_blocks, block_size, k_value)
    decoded_audio_data = reconstruct_audio(decoded_audio_blocks, k_value)
    
    # Resample the decoded audio to match the original sample rate
    decoded_audio_data_resampled = resample(decoded_audio_data, len(audio_data))
    
    # Convert the data type to np.int16
    decoded_audio_data_resampled = decoded_audio_data_resampled.astype(np.int16)
    
    # Visualize the decoded audio
    visualize_audio_waveform(decoded_audio_data_resampled)
    
    # Save the resampled decoded audio
    save_decoded_audio(decoded_audio_data_resampled, sample_rate, folder_name, 'Sound1_Enc_Dec.wav')

    # Return the decoded bitstream
    decoded_bit_stream = encode_audio(decoded_audio_blocks, k_value, block_size)

    print(f"Expected Decoded Bitstream Length: {expected_decoded_bitstream_length}")
    print(f"Actual Decoded Bitstream Length: {len(decoded_bit_stream)}")

    # Check bitstream consistency
    check_bitstream_consistency(bit_stream, decoded_bit_stream)

    return decoded_audio_data_resampled, decoded_bit_stream

def check_sample_rate(bit_stream, audio_blocks, block_size, k_value):
    # Calculate the expected sample rate
    expected_sample_rate = len(bit_stream) / (len(audio_blocks) * block_size * (k_value + 1))
    
    return expected_sample_rate

def calculate_num_sample_rates(bit_stream, block_size, k_value):
    # Calculate the number of sample rates
    total_bits = len(bit_stream)
    bits_per_block = (k_value + 1) * block_size
    num_sample_rates = total_bits / bits_per_block
    return int(num_sample_rates)

def check_bitstream_consistency(original_bit_stream, processed_bit_stream):
    if original_bit_stream == processed_bit_stream:
        print("Bitstream consistency: The bitstreams match.")
    else:
        print("Bitstream consistency: The bitstreams do not match.")

# def main():
#     # Define file and folder names
#     folder_name = 'file'
#     file_name = 'Sound1.wav'
#     encoded_file_name = 'Sound1_Enc.ex2'

#     # Load audio file and get sample rate and audio data
#     sample_rate, audio_data = load_audio_file(os.path.join(folder_name, file_name))

#     if sample_rate is None or audio_data is None:
#         return  # Exit the function if loading fails

#     # Define encoding parameters
#     block_size = 1024  # You can adjust this based on your preferences
#     k_value = 4  # You can change this to 2 bits or any other value

#     # Split audio data into blocks
#     audio_blocks = split_audio_into_blocks(audio_data, block_size)

#     # Visualize the audio data split into blocks
#     visualize_audio_blocks(audio_data, audio_blocks, block_size)

#     # Encode and save audio data
#     encode_and_save_audio(audio_blocks, k_value, folder_name, file_name)

#     # Load the encoded file
#     encoded_file_path = os.path.join(folder_name, encoded_file_name)
#     bit_stream = load_encoded_file(encoded_file_path)

#     # Print information about the original audio
#     print("\nOriginal File Information:")
#     print(f"Audio file '{file_name}' loaded successfully.")
#     print(f"Sample Rate: {sample_rate}")
#     print(f"Audio Data Shape: {audio_data.shape}")
#     num_samples = len(audio_data)
#     print(f"Number of audio samples: {num_samples}")
#     if audio_data.shape[0] == 2:
#         print("Audio is stereo.")
#     else:
#         print("Audio is mono.")

#     # Print information about the encoded data
#     print("\nEncoded File Information:")
#     print(f"Encoded data saved to '{os.path.join(folder_name, encoded_file_name)}'.")
#     print(f"Number of blocks: {len(audio_blocks)}")

#     if bit_stream is not None:
#         # Calculate the number of sample rates in the encoded data
#         num_sample_rates = calculate_num_sample_rates(bit_stream, block_size, k_value)

#         # Check sample rate consistency
#         if sample_rate != check_sample_rate(bit_stream, audio_blocks, block_size, k_value):
#             print("Sample rates do not match.")
#         else:
#             process_encoded_audio(bit_stream, audio_blocks, block_size, k_value, sample_rate, folder_name, audio_data)


#         decoded_audio_blocks = process_encoded_audio(bit_stream, audio_blocks, block_size, k_value, sample_rate, folder_name, audio_data)

#         # Check for lossless encoding
#         if np.array_equal(audio_data, decoded_audio_blocks):
#             print("Encoding and decoding is lossless.")
#         else:
#             print("Encoding and decoding is not lossless. There may be artifacts.")



#     # Load and display information about the decoded audio file
#     decoded_file_name = 'Sound1_Enc_Dec.wav'
#     decoded_file_path = os.path.join(folder_name, decoded_file_name)
#     decoded_sample_rate, decoded_audio_data = load_audio_file(decoded_file_path)
    
#     if decoded_sample_rate is not None and decoded_audio_data is not None:
#         print("\nDecoded File Information:")
#         print(f"Decoded Audio file '{decoded_file_name}' loaded successfully.")
#         print(f"Sample Rate: {decoded_sample_rate}")
#         print(f"Decoded Audio Data Shape: {decoded_audio_data.shape}")
#         num_decoded_samples = len(decoded_audio_data)
#         print(f"Number of decoded audio samples: {num_decoded_samples}")
#     else:
#         print("Unable to load the decoded audio file.")

# if __name__ == "__main__":
#     main()

def main():
    # Define file and folder names
    folder_name = 'file'
    file_name = 'Sound1.wav'
    encoded_file_name = 'Sound1_Enc.ex2'

    # Load audio file and get sample rate and audio data
    sample_rate, audio_data = load_audio_file(os.path.join(folder_name, file_name))

    if sample_rate is None or audio_data is None:
        return  # Exit the function if loading fails

    # Define encoding parameters
    block_size = 1024  # You can adjust this based on your preferences
    k_value = 4  # You can change this to 2 bits or any other value

    # Split audio data into blocks
    audio_blocks = split_audio_into_blocks(audio_data, block_size)

    # Visualize the audio data split into blocks
    visualize_audio_blocks(audio_data, audio_blocks, block_size)

    # Encode and save audio data
    encode_and_save_audio(audio_blocks, k_value, block_size, folder_name, file_name)



    # Load the encoded file
    encoded_file_path = os.path.join(folder_name, encoded_file_name)
    bit_stream = load_encoded_file(encoded_file_path)

    # Print information about the original audio
    print("\nOriginal File Information:")
    print(f"Audio file '{file_name}' loaded successfully.")
    print(f"Sample Rate: {sample_rate}")
    print(f"Audio Data Shape: {audio_data.shape}")
    num_samples = len(audio_data)
    print(f"Number of audio samples: {num_samples}")
    if audio_data.shape[0] == 2:
        print("Audio is stereo.")
    else:
        print("Audio is mono.")

    # Print information about the encoded data
    print("\nEncoded File Information:")
    print(f"Encoded data saved to '{os.path.join(folder_name, encoded_file_name)}'.")
    print(f"Number of blocks: {len(audio_blocks)}")

    if bit_stream is not None:
        # Calculate the number of sample rates in the encoded data
        num_sample_rates = calculate_num_sample_rates(bit_stream, block_size, k_value)

        # Check sample rate consistency
        if sample_rate != check_sample_rate(bit_stream, audio_blocks, block_size, k_value):
            print("Sample rates do not match.")
        else:
            process_encoded_audio(bit_stream, audio_blocks, block_size, k_value, sample_rate, folder_name, audio_data)

        decoded_audio_blocks = process_encoded_audio(bit_stream, audio_blocks, block_size, k_value, sample_rate, folder_name, audio_data)

        # Check for lossless encoding
        if np.array_equal(audio_data, decoded_audio_blocks):
            print("Encoding and decoding is lossless.")
        else:
            print("Encoding and decoding is not lossless. There may be artifacts.")



    # Load and display information about the decoded audio file
    decoded_file_name = 'Sound1_Enc_Dec.wav'
    decoded_file_path = os.path.join(folder_name, decoded_file_name)
    decoded_sample_rate, decoded_audio_data = load_audio_file(decoded_file_path)
    
    if decoded_sample_rate is not None and decoded_audio_data is not None:
        print("\nDecoded File Information:")
        print(f"Decoded Audio file '{decoded_file_name}' loaded successfully.")
        print(f"Sample Rate: {decoded_sample_rate}")
        print(f"Decoded Audio Data Shape: {decoded_audio_data.shape}")
        num_decoded_samples = len(decoded_audio_data)
        print(f"Number of decoded audio samples: {num_decoded_samples}")
    else:
        print("Unable to load the decoded audio file.")

if __name__ == "__main__":
    main()

```

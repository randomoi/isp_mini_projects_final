```python
import os
import numpy as np
from scipy.io import wavfile
import pandas as pd
import matplotlib.pyplot as plt
from tabulate import tabulate
```

```python
def visualize_waveforms(original_data, decoded_data, rate, method=None, k=None, file_name=""):
    time = np.linspace(0., len(original_data) / rate, len(original_data))
    
    # Extract only the base name from the file path for cleaner titles
    base_name = os.path.basename(file_name)

    # Deciding title based on method and k-value
    if method == "rice":
        original_title = f"{base_name} - Original Waveform with K={k}"
        decoded_title = f"{base_name} - Decoded Waveform with K={k}"
        comparison_title = f"{base_name} - Waveform Comparison with K={k}"
    else:
        original_title = f"{base_name} - Original Waveform"
        decoded_title = f"{base_name} - Decoded Waveform ({method.capitalize()})"
        comparison_title = f"{base_name} - Overlay: Original vs {method.capitalize()} Decoded Waveform"

    # Plotting the original waveform
    plt.figure(figsize=(8, 2))  # Adjusting the figure size
    plt.plot(time, original_data, marker=None)  # Remove markers
    plt.title(original_title)
    plt.xlabel("Time [s]")
    plt.ylabel("Amplitude")
    plt.show()

    # Plotting the decoded waveform
    plt.figure(figsize=(8, 2))  # Adjusting the figure size
    plt.plot(time, decoded_data, color="red", marker=None)  # Remove markers
    plt.title(decoded_title)
    plt.xlabel("Time [s]")
    plt.ylabel("Amplitude")
    plt.show()

    # Plotting both waveforms on top of each other for comparison
    plt.figure(figsize=(8, 2))  # Adjusting the figure size
    plt.plot(time, original_data, marker=None)  # Remove markers
    plt.plot(time, decoded_data, alpha=0.7, marker=None)  # Remove markers
    plt.title(comparison_title)
    plt.xlabel("Time [s]")
    plt.ylabel("Amplitude")
    plt.show()
```

## Rice Algorithm

```python
def add_padding(encoded_data):
    # Calculate how many bits are needed as padding to make the total length a multiple of 8
    padding = 8 - len(encoded_data) % 8
    padding = padding % 8  # Ensuring that if it's a multiple of 8, we don't add unnecessary padding
    # Create padding string
    padding_str = '0' * padding
    # First byte is padding info
    padding_info = format(padding, '08b')
    return padding_info + encoded_data + padding_str

def remove_padding(encoded_data_with_padding):
    # Extract padding info
    padding = int(encoded_data_with_padding[:8], 2)
    return encoded_data_with_padding[8:-padding] if padding != 0 else encoded_data_with_padding[8:]

def ensure_directory_exists(directory):
    if not os.path.exists(directory):
        os.makedirs(directory)

# Modify save_encoded_data to save in the R123 directory
def save_encoded_data(filename, encoded_data):
    ensure_directory_exists('R123')
    encoded_filepath = os.path.join('R123', os.path.basename(filename))
    
    encoded_data_with_padding = add_padding(encoded_data)
    byte_representation = int(encoded_data_with_padding, 2).to_bytes((len(encoded_data_with_padding) + 7) // 8, 'big')
    with open(encoded_filepath, 'wb') as f:
        f.write(byte_representation)

# Modify load_encoded_data to load from the R123 directory
def load_encoded_data(filename):
    encoded_filepath = os.path.join('R123', os.path.basename(filename))
    
    with open(encoded_filepath, 'rb') as f:
        byte_data = f.read()
        encoded_data_with_padding = bin(int.from_bytes(byte_data, 'big'))[2:].rjust(len(byte_data) * 8, '0')
    return remove_padding(encoded_data_with_padding)

def rice_encode(data, k):
    quotient, remainder = divmod(data, 2**k)
    unary_part = '1' * quotient + '0'
    binary_part = format(remainder, '0' + str(k) + 'b')
    return unary_part + binary_part

def rice_decode(code, k):
    unary_part = code.find('0')
    quotient = unary_part
    remainder = int(code[unary_part+1:unary_part+k+1], 2)
    return quotient * (2**k) + remainder

def encode_audio(audio_data, k):
    offset = 32768  # For 16-bit PCM WAV
    return ''.join([rice_encode(sample + offset, k) for sample in audio_data])

def decode_audio(encoded_data, k):
    offset = 32768  # For 16-bit PCM WAV
    i = 0
    decoded_samples = []
    while i < len(encoded_data):
        unary_end = encoded_data.find('0', i)
        sample = rice_decode(encoded_data[i:unary_end+k+1], k) - offset
        decoded_samples.append(sample)
        i = unary_end + k + 1
    return np.array(decoded_samples, dtype=np.int16)

def process_audio_with_rice(audio_files, k_values=[2, 4], sample_size=10):
    # Ensure directory exists
    if not os.path.exists('R123'):
        os.makedirs('R123')

    for audio_filename in audio_files:
        table_data = []  # Reset table data for each audio file

        # Read WAV
        rate, data = wavfile.read(audio_filename)
        
        for k in k_values:
            # Encode data
            encoded_data = encode_audio(data, k)
            
            # Save encoded data
            encoded_filepath = os.path.join('R123', os.path.basename(audio_filename.replace('.wav', '_Enc_K' + str(k) + '.ex2')))
            with open(encoded_filepath, 'wb') as f:
                pickle.dump(encoded_data, f)
            
            # Read encoded data
            with open(encoded_filepath, 'rb') as f:
                read_encoded_data = pickle.load(f)
            
            # Decode the data
            decoded_data = decode_audio(read_encoded_data, k)
            
            # Determine if samples match
            match_status = "Match" if np.array_equal(data, decoded_data) else "Mismatch"
            
            # Save decoded data
            decoded_filepath = os.path.join('R123', os.path.basename(audio_filename.replace('.wav', '_Enc_Dec_K' + str(k) + '.wav')))
            wavfile.write(decoded_filepath, rate, decoded_data)
            
            # Append results to table data for the current audio file
            table_data.append({
                "File": os.path.basename(audio_filename),
                "K Value": k,
                "Original Sample Count": len(data),
                "Decoded Sample Count": len(decoded_data),
                "Match Status for all Samples": match_status,
                "First 10 Original Samples": data[:sample_size],
                "First 10 Decoded Samples": decoded_data[:sample_size]
            })
            
        # Convert to DataFrame, apply styling, and display for the current audio file
        df = pd.DataFrame(table_data)
        styled_df = df.style.format({
            "Original Sample Count": "{:,}",
            "Decoded Sample Count": "{:,}",
            "First 10 Original Samples": lambda x: ", ".join(map(str, x)),
            "First 10 Decoded Samples": lambda x: ", ".join(map(str, x)),
        }).hide(axis="index")
        display(styled_df)
        
        # Visualize the waveforms for all k values after displaying the table for the current audio file
        for k in k_values:
            visualize_waveforms(data, decode_audio(encode_audio(data, k), k), rate, method="rice", k=k, file_name=audio_filename)

audio_files = ['file/Sound1.wav', 'file/Sound2.wav']
process_audio_with_rice(audio_files)
```

### Rice Algorithm Table

```python
def analyze_compression(audio_files, k_values=[2, 4]):
    rows = []
    
    for file in audio_files:
        original_size = os.path.getsize(file)
        
        row = {'File': os.path.basename(file), 'Original Size': original_size}
        
        for k in k_values:
            encoded_file_path = file.replace('file/', 'R123/').replace('.wav', f'_Enc_K{k}.ex2')
            encoded_size = os.path.getsize(encoded_file_path)



            compression_ratio = encoded_size / original_size
            compression_percent = (1 - compression_ratio) * 100
            
            # Modify the compression_percent string if it's negative
            if compression_percent < 0:
                compression_string = f"0.00% ({compression_percent:.2f}%)"
            else:
                compression_string = f"{compression_percent:.2f}%"
            
            row[f"Rice (K = {k} bits)"] = encoded_size
            row[f"% Compression (K = {k} bits)"] = compression_string
            
        rows.append(row)
    
    # Define column order
    column_order = ['File', 'Original Size'] + \
                   [f"Rice (K = {k} bits)" for k in reversed(k_values)] + \
                   [f"% Compression (K = {k} bits)" for k in reversed(k_values)]
    
    df = pd.DataFrame(rows, columns=column_order)
    return df

def human_readable_size(size, decimal_places=2):
    # Note: Now using powers of ten
    units = [10**0, 10**3, 10**6, 10**9, 10**12]  # B, KB, MB, GB, TB
    labels = ['B', 'KB', 'MB', 'GB', 'TB']
    
    for idx, unit in enumerate(units):
        if size < unit * 1000:
            return f"{size/unit:.{decimal_places}f} {labels[idx]}"
    
    return f"{size/10**12:.{decimal_places}f} TB"

audio_files = ['file/Sound1.wav', 'file/Sound2.wav']
df = analyze_compression(audio_files)

# Apply formatting using the style method
styled_df = df.style.format({
    'Original Size': human_readable_size,
    'Rice (K = 2 bits)': human_readable_size,
    'Rice (K = 4 bits)': human_readable_size
}).hide(axis="index")

display(styled_df)
```

```python
import os
import numpy as np
from scipy.io import wavfile
import matplotlib.pyplot as plt
import heapq
import pickle
from collections import Counter
import heapq

class HuffmanNode:
    def __init__(self, value, freq):
        self.value = value
        self.freq = freq
        self.left = None
        self.right = None

    def __lt__(self, other):
        return self.freq < other.freq


def build_huffman_tree(freq_table):
    heap = [HuffmanNode(value, freq) for value, freq in freq_table.items()]
    heapq.heapify(heap)
    
    while len(heap) > 1:
        left = heapq.heappop(heap)
        right = heapq.heappop(heap)
        
        merged_node = HuffmanNode(None, left.freq + right.freq)
        merged_node.left = left
        merged_node.right = right

        heapq.heappush(heap, merged_node)
    
    return heap[0]  # root node


def generate_huffman_codes(root):
    def _generate_huffman_codes(node, code, mapping):
        if node is not None:
            if node.value is not None:
                mapping[node.value] = code
            _generate_huffman_codes(node.left, code + '0', mapping)
            _generate_huffman_codes(node.right, code + '1', mapping)

    mapping = {}
    _generate_huffman_codes(root, '', mapping)
    return mapping


def huffman_encode(data, huff_codes):
    return ''.join([huff_codes[val] for val in data])


def huffman_decode(encoded_data, huff_codes):
    reversed_codes = {code: val for val, code in huff_codes.items()}
    decoded_data = []
    buffer = ''
    
    for bit in encoded_data:
        buffer += bit
        if buffer in reversed_codes:
            decoded_data.append(reversed_codes[buffer])
            buffer = ''
    return np.array(decoded_data, dtype=np.int16)


def process_audio_with_huffman(audio_files, sample_size=10):
    # Ensure directory exists
    if not os.path.exists('H123'):
        os.makedirs('H123')
    
    for audio_filename in audio_files:
        table_data = []  # Reset table data for each audio file

        # Read WAV
        rate, original_data = wavfile.read(audio_filename)

        # Huffman Encoding
        freq_table = dict(Counter(original_data))
        huffman_tree_root = build_huffman_tree(freq_table)
        huff_codes = generate_huffman_codes(huffman_tree_root)
        encoded_data = huffman_encode(original_data, huff_codes)
        
        # Save encoded data
        encoded_filepath = os.path.join('H123', os.path.basename(audio_filename).replace('.wav', '_Huff_Encoded.ex2'))
        with open(encoded_filepath, 'wb') as f:
            pickle.dump((encoded_data, huff_codes), f)

        # Read encoded data
        with open(encoded_filepath, 'rb') as f:
            read_encoded_data, read_huff_codes = pickle.load(f)

        # Huffman Decoding
        decoded_data = huffman_decode(read_encoded_data, read_huff_codes)

        # Determine if samples match
        match_status = "Match" if np.array_equal(original_data, decoded_data) else "Mismatch"

        # Write decoded data
        decoded_filepath = os.path.join('H123', os.path.basename(audio_filename).replace('.wav', '_Huff_Decoded.wav'))
        wavfile.write(decoded_filepath, rate, decoded_data)

        # Append the results for this audio file to the table data
        table_data.append({
                "File": os.path.basename(audio_filename),
                "Original Sample Count": len(original_data),
                "Decoded Sample Count": len(decoded_data),
                "Match Status for all Samples": match_status,
                "First 10 Original Samples": original_data[:sample_size],
                "First 10 Decoded Samples": decoded_data[:sample_size]
        })

        # Convert to DataFrame, apply styling, and display for the current audio file
        df = pd.DataFrame(table_data)
        styled_df = df.style.format({
            "Original Sample Count": "{:,}",
            "Decoded Sample Count": "{:,}",
            "First 10 Original Samples": lambda x: ", ".join(map(str, x)),
            "First 10 Decoded Samples": lambda x: ", ".join(map(str, x)),
        }).hide(axis="index")
        display(styled_df)

        # Visualize the waveforms for the current audio file
        visualize_waveforms(original_data, decoded_data, rate, method="huffman", file_name=audio_filename)

# Run for each audio file
audio_files = ['file/Sound1.wav', 'file/Sound2.wav']
process_audio_with_huffman(audio_files)
```

## Compare Rice, Huffman and FLAC

```python
import os
import pandas as pd

def human_readable_size(size):
    """Convert byte size to a human readable format."""
    for unit in ['B','KB','MB','GB','TB']:
        if size < 1024.0:
            return f"{size:3.2f} {unit}"
        size /= 1024.0

def analyze_compression(audio_files, k_values=[2, 4]):
    rows = []

    for file in audio_files:
        original_size = os.path.getsize(file)
        
        row = {'File': os.path.basename(file), 'Original Size': original_size}
        
        # FLAC compression results
        flac_file_path = os.path.join("F123", file.replace('.wav', '.flac').replace("file/", ""))
        flac_size = os.path.getsize(flac_file_path)
        flac_compression_ratio = flac_size / original_size
        flac_compression_percent = (1 - flac_compression_ratio) * 100
        if flac_compression_percent < 0:
            flac_compression_string = f"0.00% ({flac_compression_percent:.2f}%)"
        else:
            flac_compression_string = f"{flac_compression_percent:.2f}%"

        row['FLAC Encoded Size'] = flac_size
        row['% Compression (FLAC)'] = flac_compression_string
   
        for k in k_values:
            # Update the path for Rice encoded files
            rice_encoded_path = os.path.join("R123", file.replace('.wav', '_Enc_K' + str(k) + '.ex2').replace("file/", ""))
            encoded_size = os.path.getsize(rice_encoded_path)
            compression_ratio = encoded_size / original_size
            compression_percent = (1 - compression_ratio) * 100
            
            if compression_percent < 0:
                compression_string = f"0.00% ({compression_percent:.2f}%)"
            else:
                compression_string = f"{compression_percent:.2f}%"
            
            row[f"Rice (K = {k} bits)"] = encoded_size
            row[f"% Compression (K = {k} bits)"] = compression_string
        
        # Update the path for Huffman encoded files
        huffman_encoded_path = os.path.join("H123", file.replace('.wav', '_Huff_Encoded.ex2').replace("file/", ""))
        huffman_encoded_size = os.path.getsize(huffman_encoded_path)
        huffman_compression_ratio = huffman_encoded_size / original_size
        huffman_compression_percent = (1 - huffman_compression_ratio) * 100
        if huffman_compression_percent < 0:
            huffman_compression_string = f"0.00% ({huffman_compression_percent:.2f}%)"
        else:
            huffman_compression_string = f"{huffman_compression_percent:.2f}%"
        
        row['Huffman Encoded Size'] = huffman_encoded_size
        row['% Compression (Huffman)'] = huffman_compression_string
        
        rows.append(row)

    # Define column order
    column_order = ['File', 'Original Size', 
                    'Rice (K = 2 bits)', 'Rice (K = 4 bits)', 
                    'Huffman Encoded Size', 
                    'FLAC Encoded Size', 
                    '% Compression (K = 2 bits)', '% Compression (K = 4 bits)', 
                    '% Compression (Huffman)', '% Compression (FLAC)']

    df = pd.DataFrame(rows, columns=column_order)
    return df

# Let's run the function and display the table
audio_files = ['file/Sound1.wav', 'file/Sound2.wav']
df = analyze_compression(audio_files)

# Apply formatting using the style method
styled_df = df.style.format({
    'Original Size': human_readable_size,
    'Rice (K = 2 bits)': human_readable_size,
    'Rice (K = 4 bits)': human_readable_size,
    'Huffman Encoded Size': human_readable_size,
    'FLAC Encoded Size': human_readable_size,
}).hide(axis="index")

display(styled_df)
```

```python
import os
from scipy.io import wavfile
import numpy as np
import bitarray
import matplotlib.pyplot as plt
from scipy.signal import resample 
import json

def load_audio_file(file_path):
    try:
        sample_rate, audio_data = wavfile.read(file_path)
        return sample_rate, audio_data
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' does not exist.")
        return None, None
    except Exception as e:
        print(f"An error occurred while loading the audio file: {str(e)}")
        return None, None
    
def split_audio_into_blocks(audio_data, block_size):
    num_blocks = len(audio_data) // block_size
    blocks = np.array_split(audio_data, num_blocks)
    return blocks

def rice_encode(value, k_value):
    q, r = divmod(value, 2**k_value)
    unary_code = '1' * q + '0'
    binary_code = format(r, '0' + str(k_value) + 'b')
    encoded_value = unary_code + binary_code
    return encoded_value

def rice_decode(encoded_value, k_value):
    idx = encoded_value.index('0')
    q = idx
    binary_code = encoded_value[idx + 1 : idx + 1 + k_value]
    r = int(binary_code, 2)
    decoded_value = q * 2**k_value + r
    return decoded_value

def encode_audio(audio_data, k_value, block_size):
    encoded_blocks = []
    for i in range(0, len(audio_data), block_size):
        block = audio_data[i:i + block_size]
        encoded_block = [rice_encode(value, k_value) for value in block]
        encoded_blocks.append(encoded_block)
    return encoded_blocks

def decode_audio(bit_stream, k_value):
    decoded_audio_data = []
    bit_stream_index = 0

    while bit_stream_index < len(bit_stream):
        unary_code = ""

        # Read unary code
        while bit_stream_index < len(bit_stream) and bit_stream[bit_stream_index]:
            unary_code += "1"
            bit_stream_index += 1
        
        if bit_stream_index < len(bit_stream):
            unary_code += "0"
            bit_stream_index += 1

            # Check if there are enough bits for the binary code
            if bit_stream_index + k_value <= len(bit_stream):
                binary_code = bit_stream[bit_stream_index:bit_stream_index + k_value]
                bit_stream_index += k_value
            else:
                # Handle incomplete binary code at the end of the bitstream
                break

            # Convert unary and binary code to value
            q = len(unary_code) - 1
            r = int(binary_code.to01(), 2)
            decoded_value = q * (2**k_value) + r
            decoded_audio_data.append(decoded_value)
        else:
            break

    return np.array(decoded_audio_data)


def decode_audio_blocks(bit_stream, audio_blocks, block_size, k_value):
    decoded_audio_blocks = []
    bit_stream_index = 0

    for _ in range(len(audio_blocks)):
        decoded_block = []

        for _ in range(block_size):
            encoded_value_str = bit_stream[bit_stream_index:bit_stream_index + (k_value + 1)].to01()
            
            if encoded_value_str:  # Check if it's not an empty string
                encoded_value = int(encoded_value_str, 2)
                decoded_block.append(encoded_value)
            
            bit_stream_index += (k_value + 1)

        decoded_audio_blocks.append(decoded_block)

    return decoded_audio_blocks


def check_encoded_data(encoded_file_path, k_value):
    try:
        with open(encoded_file_path, 'rb') as encoded_file:
            bit_stream = bitarray.bitarray()
            bit_stream.fromfile(encoded_file)
    except FileNotFoundError:
        print(f"Error: The encoded file '{encoded_file_path}' does not exist.")
        return None

    print(f"Bitstream Length: {len(bit_stream)}")

    # Decode audio data block by block
    decoded_audio_blocks = []
    bit_stream_index = 0

    # Decode audio data block by block using the decode_audio function
    decoded_audio_data = decode_audio(bit_stream, k_value)

    return decoded_audio_data

def visualize_audio_waveform(decoded_audio_data):
    if decoded_audio_data is not None:
        plt.figure(figsize=(10, 4))
        plt.plot(decoded_audio_data)
        plt.title("Decoded Audio Waveform")
        plt.xlabel("Sample")
        plt.ylabel("Amplitude")
        # plt.show()

def save_sample_rate(sample_rate, encoded_file_path):
    sample_rate_file_path = os.path.splitext(encoded_file_path)[0] + "_SampleRate.json"
    with open(sample_rate_file_path, 'w') as sample_rate_file:
        json.dump({"sample_rate": sample_rate}, sample_rate_file)

def save_encoded_data(encoded_data, encoded_file_path):
    with open(encoded_file_path, 'wb') as encoded_file:
        bit_stream = bitarray.bitarray()
        encoded_audio, sample_rate = encoded_data
        for encoded_block in encoded_audio:
            for encoded_value in encoded_block:
                bit_stream.extend(encoded_value)
        bit_stream.tofile(encoded_file)
        
    # Save the sample rate as metadata in a separate JSON file
    save_sample_rate(sample_rate, encoded_file_path)


def load_encoded_file(encoded_file_path):
    try:
        with open(encoded_file_path, 'rb') as encoded_file:
            bit_stream = bitarray.bitarray()
            bit_stream.fromfile(encoded_file)
        return bit_stream
    except FileNotFoundError:
        print(f"Error: The encoded file '{encoded_file_path}' does not exist.")
        return None
    except Exception as e:
        print(f"An error occurred while loading the encoded file: {str(e)}")
        return None


def reconstruct_audio(decoded_audio_blocks, k_value):
    decoded_audio_data = decode_audio(decoded_audio_blocks, k_value)
    return decoded_audio_data

def save_decoded_audio(decoded_audio_data, sample_rate, folder_name, file_name):
    decoded_file_name = "Sound1_Enc_Dec.wav"  # Specify the desired file name
    decoded_file_path = os.path.join(folder_name, decoded_file_name)
    
    try:
        wavfile.write(decoded_file_path, sample_rate, decoded_audio_data.astype(np.int16))
        print(f"Decoded audio data saved to '{decoded_file_path}'.")
    except Exception as e:
        print(f"An error occurred while saving the decoded audio data: {str(e)}")


    
def check_sample_rate(bit_stream, audio_blocks, block_size, k_value):
    # Calculate the expected sample rate
    expected_sample_rate = len(bit_stream) / (len(audio_blocks) * block_size * (k_value + 1))
    
    return expected_sample_rate

def calculate_num_sample_rates(bit_stream, block_size, k_value):
    # Calculate the number of sample rates
    total_bits = len(bit_stream)
    bits_per_block = (k_value + 1) * block_size
    num_sample_rates = total_bits / bits_per_block
    return int(num_sample_rates)

def check_bitstream_consistency(original_bit_stream, processed_bit_stream):
    if original_bit_stream == processed_bit_stream:
        print("Bitstream consistency: The bitstreams match.")
    else:
        print("Bitstream consistency: The bitstreams do not match.")

def process_encoded_audio(bit_stream, block_size, k_value, sample_rate, folder_name, file_name, audio_data):
    # Decode audio using the provided block_size and k_value
    decoded_audio_data = decode_audio(bit_stream, k_value)

    # Resample the decoded audio to match the original sample rate
    decoded_audio_data_resampled = resample(decoded_audio_data, len(decoded_audio_data) * sample_rate // len(audio_data))
    decoded_audio_data_resampled = decoded_audio_data_resampled.astype(np.int16)

    # Visualize the decoded audio
    visualize_audio_waveform(decoded_audio_data_resampled)

    # Save the resampled decoded audio with the specified file name
    save_decoded_audio(decoded_audio_data_resampled, sample_rate, folder_name, file_name)

    # Check bitstream consistency
    check_bitstream_consistency(bit_stream, decoded_audio_data)

    # Check for lossless encoding
    if np.array_equal(audio_data, decoded_audio_data):
        print("Encoding and decoding is lossless.")
    else:
        print("Encoding and decoding is not lossless. There may be artifacts.")


def main():
    # Define file and folder names
    folder_name = 'file'
    file_name = 'Sound1.wav'
    encoded_file_name = 'Sound1_Enc.ex2'

    # Load audio file and get sample rate and audio data
    sample_rate, audio_data = load_audio_file(os.path.join(folder_name, file_name))

    if sample_rate is None or audio_data is None:
        return  # Exit the function if loading fails

    # Define encoding parameters
    block_size = 1024  # You can adjust this based on your preferences
    k_value = 4  # You can change this to 2 bits or any other value

    # Split audio data into blocks
    audio_blocks = split_audio_into_blocks(audio_data, block_size)

    # Visualize the audio data split into blocks
    visualize_audio_blocks(audio_data, audio_blocks, block_size)

    # Encode and save audio data
    # encode_and_save_audio(audio_blocks, k_value, block_size, folder_name, file_name)
    encode_and_save_audio(audio_data, k_value, block_size, folder_name, file_name, sample_rate)

    # Get the number of samples in the audio data
    num_samples = len(audio_data)

    # Load the encoded file
    encoded_file_path = os.path.join(folder_name, encoded_file_name)
    bit_stream = load_encoded_file(encoded_file_path)


    # Print information about the original audio
    print("\nOriginal File Information:")
    print(f"Audio file '{file_name}' loaded successfully.")
    print(f"Sample Rate: {sample_rate}")
    print(f"Audio Data Shape: {audio_data.shape}")
    num_samples = len(audio_data)
    print(f"Number of audio samples: {num_samples}")
    if audio_data.shape[0] == 2:
        print("Audio is stereo.")
    else:
        print("Audio is mono.")

    # Print information about the encoded data
    print("\nEncoded File Information:")
    print(f"Encoded data saved to '{os.path.join(folder_name, encoded_file_name)}'.")
    print(f"Number of blocks: {len(audio_blocks)}")



    if bit_stream is not None:
        # Process the encoded audio and visualize/save the decoded audio
        process_encoded_audio(bit_stream, block_size, k_value, sample_rate, folder_name, file_name, audio_data)  # Pass audio_data here

        # Calculate the number of sample rates in the encoded data
        num_sample_rates = calculate_num_sample_rates(bit_stream, block_size, k_value)

        decoded_audio_blocks = process_encoded_audio(bit_stream, block_size, k_value, sample_rate, folder_name, file_name, audio_data)  # Pass audio_data here


        
        # Check for lossless encoding
        if np.array_equal(audio_data, decoded_audio_blocks):
            print("Encoding and decoding is lossless.")
        else:
            print("Encoding and decoding is not lossless. There may be artifacts.")

   
    # Load and display information about the decoded audio file
    decoded_file_name = 'Sound1_Enc_Dec.wav'
    decoded_file_path = os.path.join(folder_name, decoded_file_name)
    decoded_sample_rate, decoded_audio_data = load_audio_file(decoded_file_path)
    
    if decoded_sample_rate is not None and decoded_audio_data is not None:
        print("\nDecoded File Information:")
        print(f"Decoded Audio file '{decoded_file_name}' loaded successfully.")
        print(f"Sample Rate: {decoded_sample_rate}")
        print(f"Decoded Audio Data Shape: {decoded_audio_data.shape}")
        num_decoded_samples = len(decoded_audio_data)
        print(f"Number of decoded audio samples: {num_decoded_samples}")
    else:
        print("Unable to load the decoded audio file.")

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import sounddevice as sd
from scipy.io import wavfile
import math

# RICE ENCODER
def rice_encoder(S, M, K):
    if S < 0:
        sign = "1"  # Negative
    else:
        sign = "0"  # Positive

    quotient = int(math.floor(abs(S) / M))
    remainder = abs(S) % M

    q_code_str = "1" * quotient
    diff = int(math.pow(2, K) - M)

    if remainder < diff:
        b = K - 1
        r_code_str = "{0:0{1}b}".format(remainder, b)
    else:
        b = K
        r_code_str = "{0:0{1}b}".format(remainder + diff, b)

    code_word = sign + q_code_str + "0" + r_code_str
    return code_word

# RICE DECODER
def rice_decoder(encoded_str, M, K):
    idx = 0
    sign = int(encoded_str[idx])
    idx += 1

    q_end_idx = encoded_str.find('0', idx)
    q = q_end_idx - idx
    idx = q_end_idx + 1

    diff = int(math.pow(2, K) - M)

    if int(encoded_str[idx:idx+K-1], 2) < diff:
        remainder = int(encoded_str[idx: idx + K - 1], 2)
        idx += (K - 1)
    else:
        remainder = int(encoded_str[idx: idx + K], 2) - diff
        idx += K

    value = q * M + remainder

    if sign == 1:
        value = -value

    bits_used = idx  # Here, idx now contains the total bits consumed by the decoder
    return value, bits_used


# Read the audio file to encode
filepath = 'file/Sound1.wav'
sample_rate, audio = wavfile.read(filepath)

# Check if stereo or mono
channels = 1 if len(audio.shape) == 1 else audio.shape[1]

# Encode the audio file
for K in [2, 4]:
    M = 4
    encoded_audio = ""
    for i in range(len(audio)):
        if channels == 1:
            encoded_audio += rice_encoder(audio[i], M, K)
        else:
            for j in range(channels):
                encoded_audio += rice_encoder(audio[i][j], M, K)

    # Save the encoded audio as binary
    encoded_str = ''.join(encoded_audio)
    byte_array = bytearray()
    for i in range(0, len(encoded_str), 8):
        byte_array.append(int(encoded_str[i:i+8], 2))

    with open('encoded_audio.bin', 'wb') as f:
        f.write(byte_array)

    # Decoding the saved binary to check for any mismatches
    with open('encoded_audio.bin', 'rb') as f:
        content = f.read()
        encoded_str = ''.join(bin(byte)[2:].zfill(8) for byte in content)

    # Decoding
    decoded_audio = []

    idx = 0
    remaining_samples = len(audio) * channels  # If stereo, account for both channels in samples count
    while idx < len(encoded_str) and remaining_samples > 0: 
        # Extract sign bit
        sign = encoded_str[idx]
        idx += 1

        # Extract quotient
        q_end_idx = encoded_str.find('0', idx)
        q = q_end_idx - idx
        idx = q_end_idx + 1

        # Determine K for remainder
  
        remainder = encoded_str[idx: idx + K]
        idx += K

        encoded_sample = sign + '1' * q + '0' + remainder
        decoded_value, _ = rice_decoder(encoded_sample, M, K)
        decoded_audio.append(decoded_value)

        remaining_samples -= 1

        if remaining_samples == 0:
            break

    decoded_audio = np.array(decoded_audio, dtype=np.int16)
    if channels > 1:
        decoded_audio = decoded_audio.reshape(-1, channels)

    # Play decoded audio
    # print('Playing decoded audio...!')
    # sd.play(decoded_audio, sample_rate)

    # Write decoded audio back to .wav format
    decoded_file = 'decoded_audio.wav'
    wavfile.write(decoded_file, sample_rate, decoded_audio)


    # Print results:
    print(f"Using K={K}")
    print("Original Samples:", len(audio))
    print("Decoded Samples:", len(decoded_audio))
    
# Displaying the results
# print("Encoded Sample:", final_encoded_sample)
print("Decoded Value:", decoded_value)
print("Original Value:", audio[-1])


# Displaying the results
# print("Encoded Sample:", final_encoded_sample)
print("Decoded Value:", decoded_value)
print("Original Value:", audio[-1])


print("Original Samples:", len(audio))
print("Decoded Samples:", len(decoded_audio))


import matplotlib.pyplot as plt

plt.hist(audio, bins=50)
plt.xlabel("Sample Value")
plt.ylabel("Frequency")
plt.show()


print("Last 10 original samples:", audio[-10:])
print("Last 10 decoded samples:", decoded_audio[-10:])
print("Last 50 bits of encoded string:", encoded_str[-50:])


# final_encoded_sample = rice_encoder(audio[-1], M)
# final_decoded_value = rice_decoder(final_encoded_sample, M)
# print("Final encoded sample:", final_encoded_sample)
# print("Final decoded value:", final_decoded_value)



import matplotlib.pyplot as plt

# Visualizing the original and decoded audio
plt.figure(figsize=(15, 6))

# Plot the original waveform
plt.subplot(2, 1, 1)
plt.plot(audio, 'b')
plt.title('Original Audio')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

# Plot the decoded waveform
plt.subplot(2, 1, 2)
plt.plot(decoded_audio, 'r')
plt.title('Decoded Audio')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()


import matplotlib.pyplot as plt

# Visualizing the original and decoded audio on the same plot
plt.figure(figsize=(15, 6))

# Plot the original waveform
plt.plot(audio, 'b', label='Original Audio')

# Plot the decoded waveform
plt.plot(decoded_audio, 'r', alpha=0.7, label='Decoded Audio')

plt.title('Original vs. Decoded Audio')
plt.xlabel('Sample')
plt.ylabel('Amplitude')
plt.legend()
plt.show()


import matplotlib.pyplot as plt

# Trim the decoded audio to match the length of the original audio
decoded_audio_trimmed = decoded_audio[:len(audio)]

# Calculate the difference between the two signals
difference = audio - decoded_audio_trimmed

# Plotting the difference
plt.figure(figsize=(15, 6))
plt.plot(difference, 'g', label='Difference (Original - Decoded)')
plt.title('Difference between Original and Decoded Audio')
plt.xlabel('Sample')
plt.ylabel('Difference')
plt.legend()
plt.show()


# Number of samples to inspect at the end
num_samples = 5

print("Last", num_samples, "original samples:", audio[-num_samples:])
print("Last", num_samples, "decoded samples:", decoded_audio_trimmed[-num_samples:])
print("Difference:", difference[-num_samples:])


# # Encoding the last sample
# final_encoded_sample = rice_encoder(audio[-1], M)
# print("Final encoded sample for the last audio sample:", final_encoded_sample)

# # Decoding the encoded sample
# final_decoded_value, _ = rice_decoder(final_encoded_sample, M)  # Ignoring bits_used again
# print("Decoded value for the final encoded sample:", final_decoded_value)


mismatched_indices = [i for i, (x, y) in enumerate(zip(audio, decoded_audio)) if x != y]

if mismatched_indices:
    first_mismatch_index = mismatched_indices[0]
    print(f"First mismatch at index: {first_mismatch_index}")
    print(f"Original value: {audio[first_mismatch_index]}, Decoded value: {decoded_audio[first_mismatch_index]}")
else:
    print("No mismatches found.")

test_sample = 1
encoded_test = rice_encoder(test_sample, M)
print(f"Encoded value for {test_sample}: {encoded_test}")
decoded_test, _ = rice_decoder(encoded_test, M)
print(f"Decoded value: {decoded_test}")


mismatch_idx = 501021
context_range = 0# Adjust this as needed
for i in range(mismatch_idx - context_range, min(mismatch_idx + context_range + 1, len(audio))):
    encoded_sample = rice_encoder(audio[i], M)
    print(f"Original: {audio[i]}, Encoded: {encoded_sample}, Decoded: {decoded_audio[i]}")
```
